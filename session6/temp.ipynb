```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import numpy as np
np.random.seed(32)

def xor_2d(n_samples=8000, noise=0.15):
    X = np.random.uniform(-1, 1, (n_samples, 2))
    
    # XOR structure
    y = (X[:, 0] * X[:, 1] > 0).astype(int)
    
    # add label noise
    flip = np.random.rand(n_samples) < noise
    y[flip] = 1 - y[flip]
    
    return X, y
```

```python
from sklearn.model_selection import train_test_split

X, y = xor_2d(n_samples=1000, noise=0.15)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score

models = {
    "Decision Tree (shallow)": DecisionTreeClassifier(max_depth=2),
    
    "Decision Tree (deep)": DecisionTreeClassifier(
        max_depth=25,
    ),
    
    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=1,
        random_state=42
    ),
    
    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    )
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    results[name] = accuracy_score(y_test, preds)

results
```
